## Create DOotplot.  

```r
create_dotplot <- function(srtobj, cell_type, cell_marker, res){
    ##  strobj = seurat object,  
    ##  cell_type = type of cells found in the cancer sample
    ##  res = resolution of dimplot
    ##  cell_marker = marker of cell type.  
    srtobj <- SetIdent(srtobj, value = res)
    dotplot <- DotPlot(srtobj, features = cell_marker, group.by = res) +
        RotatedAxis() +
        scale_colour_gradientn(colours = c("blue", "white", "red"), space = "Lab") +
        theme_minimal() +
        theme(
        legend.position = "right",
        axis.text.x = element_text(angle = 45, hjust = 1),
        plot.title = element_text(hjust = 0.5)  # Center the title
    )   +
        ggtitle(paste("Expression of", cell_type, "Markers Across Clusters"))


    filepath <- paste0("./out/dotplot/",res, "/", cell_type , ".png")
    directory <- dirname(filepath)

    # Create the directory if it doesn't exist
    if (!dir.exists(directory)) {
        dir.create(directory, recursive = TRUE)
    }
    ggsave(filepath , plot = dotplot, width = 10, height = 8)


}
```

```r
options(future.globals.maxSize = 3 * 1024^3)
markers <- FindAllMarkers(seurat_obj, return.thresh = 0.05, min.pct = 0.2, verbose = F)

split_markers <- split(markers, as.factor(markers$cluster))
split_markers_sort <- lapply(split_markers, function(x) {
  ordered <- order(x$avg_log2FC, decreasing = T)
  ordered_x <- x[ordered, ]
})

writexl::write_xlsx(split_markers_sort, path ="./out/markers/all_markers_res_0.1.xlsx", col_names = T)
```

```r
library(ComplexHeatmap)
library(RColorBrewer)
library(dplyr)
library(tibble)
library(tidyr)

markers_subset = subset(markers, subset=avg_log2FC>=0.25)

# Prepare heatmap matrix (genes x clusters)
heatmap_matrix <- markers_subset %>%
  filter(gene %in% unlist(cell_symbol)) %>%
  select(cluster, gene, avg_log2FC) %>%
  pivot_wider(names_from = cluster, values_from = avg_log2FC, values_fill = 0) %>%
  column_to_rownames(var = "gene")

# Create row annotations for cell types
row_annotation <- data.frame(Cell_Type = sapply(rownames(heatmap_matrix), function(g) {
  cell_type <- names(cell_symbol)[sapply(cell_symbol, function(x) g %in% x)]
  if (length(cell_type) > 0) return(cell_type) else return("Unknown")
}))
row_annotation <- rowAnnotation(Cell_Type = row_annotation$Cell_Type, 
                                 col = list(Cell_Type = c("Macrophage" = "blue", "CD4_T_Cell" = "red")))

# Generate heatmap
heatmap_colors <- colorRampPalette(rev(brewer.pal(n = 11, name = "RdBu")))(100)
p <- Heatmap(heatmap_matrix,
        name = "Log2FC",
        col = heatmap_colors,
        row_title = "Genes",
        column_title = "Clusters",
        show_row_names = TRUE,
        show_column_names = TRUE,
        top_annotation = NULL,
        left_annotation = row_annotation,
        cluster_rows = TRUE,
        cluster_columns = TRUE)

ggsave("./out/markers/heatmap_0.1.png", p)
```
