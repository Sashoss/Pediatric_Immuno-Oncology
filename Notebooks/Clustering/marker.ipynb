### Find Marker. 

```r
library(Seurat)
```

```r
CANCER_TYPE = "HGG"
```

```r
res = "RNA_snn_res.0.9"
```

```r
combined <- readRDS("./out/clustered_harmony_HGG.rds")
```

```r
combined <- SetIdent(combined, value = res)
markers <- FindAllMarkers(combined, only.pos = TRUE, min.pct = 0.25, logfc.threshold = 0.25)

# Split markers by cluster and sort by avg_log2FC
markers_by_cluster <- split(markers, markers$cluster)
markers_by_cluster <- lapply(markers_by_cluster, function(x) {
  x[order(x$avg_log2FC, decreasing = TRUE),]
})

# Save to Excel with each cluster in separate sheet
library(writexl)
write_xlsx(markers_by_cluster, path = paste0("./out/cluster_markers_res_0.1_", CANCER_TYPE, ".xlsx"))
```
