```r
# Function to generate dot plots for each T cell type
generate_dot_plot <- function(cell_type, marker_set) {
  # Ensure markers are unique to prevent errors in DotPlot
  unique_markers <- unique(marker_set)
  
  dotplot <- DotPlot(combined, features = unique_markers) +
    RotatedAxis() +
    scale_colour_gradientn(colours = c("blue", "yellow", "red"), space = "Lab") +
    theme_minimal() +
    theme(
      legend.position = "right",
      axis.text.x = element_text(angle = 45, hjust = 1),
      plot.title = element_text(hjust = 0.5)  # Center the title
    ) +
    ggtitle(paste("Expression of", cell_type, "Markers Across Clusters"))
  
  # Print and save the plot

  ggsave(paste0("./out/dotplots/",cell_type, "_markers_dotplot_0.1.png"), plot = dotplot, width = 10, height = 8)
}

# Generate plots for each T cell type using the updated function
for (cell_type in names(markers)) {
  generate_dot_plot(cell_type, markers[[cell_type]])
}
```

```r
# Combine markers with corresponding cell type labels
marker_groups <- data.frame(
  Gene = c(macrophage_markers, tcell_markers, nk_markers, b_cell_markers),
  CellType = c(
    rep("Macrophage", length(macrophage_markers)),
    rep("T Cell", length(tcell_markers)),
    rep("NK Cell", length(nk_markers)),
    rep("B Cell", length(b_cell_markers))
  )
)

# Create a new label combining CellType and Gene
marker_groups$Label <- paste(marker_groups$CellType, marker_groups$Gene, sep = " | ")

# Ensure markers are ordered properly for display
marker_groups <- marker_groups %>% arrange(CellType)

# Set the factor levels for Label to ensure correct ordering
marker_groups$Label <- factor(marker_groups$Label, levels = marker_groups$Label)

# Create a heatmap with combined labels for y-axis
heatmap_plot <- DoHeatmap(
  object = combined_new, 
  features = marker_groups$Gene, # Use genes for plotting
  group.by = "RNA_snn_res.0.4",  # Adjust to your clustering column
  label = TRUE, 
  size = 5
) + 
  scale_fill_viridis_c(option = "inferno") +  # Apply color scheme
  theme_minimal(base_size = 14) +             # Enhance theme aesthetics
  scale_y_discrete(labels = marker_groups$Label) +  # Set custom y-axis labels
  theme(
    axis.text.y = element_text(size = 10),    # Adjust Y-axis text size for combined labels
    axis.text.x = element_text(angle = 45, hjust = 1, size = 10) # Rotate X-axis labels
  )

# Save the heatmap to file
ggsave("./out/combined_heatmap_with_labels.png", heatmap_plot, width = 10, height = 8)

# Display the heatmap
heatmap_plot
```
