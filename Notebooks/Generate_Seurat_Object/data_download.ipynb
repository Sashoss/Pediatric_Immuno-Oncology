### download the GEO ids. 

```python
%%bash
#!/bin/bash

# Create the 'in' directory if it doesn't exist
mkdir -p in
# List of GEO IDs
ids=(GSE231859 GSE210568 GSE163656 GSE163655 GSE244433 GSE222850)
# Loop through each GEO ID
for id in "${ids[@]}"; do
    echo "Processing $id..."
    mkdir -p in/"$id"
    # Download the tar file into the subdirectory
    wget -O in/"$id"/"$id".tar "https://www.ncbi.nlm.nih.gov/geo/download/?acc=$id&format=file"
    if [[ $? -eq 0 ]]; then
        echo "Extracting $id..."
        tar -xvf in/"$id"/"$id".tar -C in/"$id"
        rm in/"$id"/"$id".tar
    else
        echo "Failed to download $id. Skipping extraction."
    fi
done
echo "All processes complete."
```

### Decompress them. 

```python
%%sh
#!/bin/bash

 # Base directory containing GSE folders
BASE_DIR="in"
# Loop through each GSE folder in the base directory
for gse_dir in "$BASE_DIR"/GSE*; do
    if [[ -d "$gse_dir" ]]; then
        echo "Processing directory: $gse_dir"
        # Find all .gz files in the directory and decompress them
        find "$gse_dir" -name "*.tar" -exec gunzip {} \;
    fi

done
echo "Decompression complete."
```
