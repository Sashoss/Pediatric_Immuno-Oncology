```r
library(DoubletFinder)
library(Seurat)
# List RDS files in the output directory
rds_files <- list.files("../Generate_Seurat_Object/out", pattern = "\\.rds$", full.names = TRUE)

# Read each RDS file into a list
seurat_objects <- list()
for (file in rds_files) {
    sample_name <- basename(file)
    sample_name <- gsub("\\.rds$", "", sample_name)
    seurat_obj <- readRDS(file)
    seurat_objects[[sample_name]] <- seurat_obj 
    print(file)
    print(colnames(seurat_obj@meta.data))
    cat("-----------------------------------\n")
    #print(sample_name, unique(seurat_obj@meta.data))
}
```

```r
# Process each Seurat object for doublet detection
for (sample_name in names(seurat_objects)) {
    print(seurat_objects[[sample_name]])
    # Normalize and find variable features
    suppressMessages({
        seurat_objects[[sample_name]] <- NormalizeData(seurat_objects[[sample_name]])
        seurat_objects[[sample_name]] <- FindVariableFeatures(seurat_objects[[sample_name]])
        seurat_objects[[sample_name]] <- ScaleData(seurat_objects[[sample_name]])
        
        # Calculate number of PCs based on dataset size
        n_features <- nrow(seurat_objects[[sample_name]])
        n_cells <- ncol(seurat_objects[[sample_name]])
        num_pcs <- 20
        
        # Run PCA with adjusted number of PCs
        seurat_objects[[sample_name]] <- RunPCA(seurat_objects[[sample_name]], npcs = num_pcs)
        
        # Run UMAP and clustering
        seurat_objects[[sample_name]] <- RunUMAP(seurat_objects[[sample_name]], dims = 1:num_pcs)
        seurat_objects[[sample_name]] <- FindNeighbors(seurat_objects[[sample_name]], dims = 1:num_pcs)
        seurat_objects[[sample_name]] <- FindClusters(seurat_objects[[sample_name]], resolution = 0.5)
        
        # pK Identification
        sweep.res <- paramSweep(seurat_objects[[sample_name]], PCs = 1:num_pcs, sct = FALSE)
        sweep.stats <- summarizeSweep(sweep.res, GT = FALSE)
        bcmvn <- find.pK(sweep.stats)
        pk_value <- as.numeric(as.character(bcmvn$pK[which.max(bcmvn$BCmetric)]))
         
        # Estimate number of doublets based on 10x guidelines (rule of thumb)
        nExp_poi <- round(0.08 * ncol(seurat_objects[[sample_name]])) 
        
        # Run DoubletFinder
        seurat_objects[[sample_name]] <- doubletFinder(
            seu = seurat_objects[[sample_name]], 
            PCs = 1:num_pcs, 
            pN = 0.25, 
            pK = pk_value, 
            nExp = nExp_poi,
            reuse.pANN = FALSE,
            sct = FALSE
        )
    })
}
```

```r
for (sample_name in names(seurat_objects)) {
    # Standardize column names
    # Find the DF.classifications column (it has a dynamic name)
    df_class_col <- grep("DF.classifications", colnames(seurat_objects[[sample_name]]@meta.data), value = TRUE)
    df_score_col <- grep("pANN", colnames(seurat_objects[[sample_name]]@meta.data), value = TRUE)
    
    # Rename columns to standardized names
    seurat_objects[[sample_name]]@meta.data$doublet_label <- seurat_objects[[sample_name]]@meta.data[[df_class_col]]
    seurat_objects[[sample_name]]@meta.data$doublet_score <- seurat_objects[[sample_name]]@meta.data[[df_score_col]]
    
    # Remove original columns
    seurat_objects[[sample_name]]@meta.data[[df_class_col]] <- NULL
    seurat_objects[[sample_name]]@meta.data[[df_score_col]] <- NULL
    
    print(paste("Completed doublet detection for", sample_name))
}

```

```r
# Save each Seurat object as an RDS file
for (sample_name in names(seurat_objects)) {
    # Create output filename
    output_file <- paste0("./out/", sample_name, "_doublets.rds")
    
    # Save the Seurat object
    saveRDS(seurat_objects[[sample_name]], file = output_file)
    
    print(paste("Saved", sample_name, "to", output_file))
}
```
