```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = open("./in/phosphorylation_results.txt")
```

```python
store_phospho = dict()
for line in data:
    if "# human" in line:
        lobj = [x for x in line.strip().split(" ") if x != ""]
        if lobj[6]=="unsp":
            continue 

        try:
            store_phospho[lobj[6]] 
        except:
            store_phospho[lobj[6]] = dict()

        store_phospho[lobj[6]][int(lobj[2])] = float(lobj[5])
```

```python
df = pd.DataFrame(store_phospho)
df
```

```python
df_filled = df.fillna(0)
```

```python
annot = df_filled.applymap(lambda x: "*" if pd.notnull(x) and x > 0.8 else "")
annot
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap, BoundaryNorm, LinearSegmentedColormap

df_sorted = df_filled.sort_index()
df_sorted = df_sorted.sort_index(axis=1)

annot = df_sorted.applymap(lambda x: "*" if x > 0.8 else "")
colors = [(1, 1, 1)] + sns.color_palette("BuPu", 256).as_hex()  
cmap = ListedColormap(colors)
norm = plt.Normalize(vmin=0.5, vmax=1.0)

plt.figure(figsize=(10, 20))
ax = sns.heatmap(df_sorted, 
                 annot=annot,
                 fmt="", 
                 linewidths=0,
                 linecolor=None,
                 cmap=cmap,
                 norm=norm,
                 cbar_kws={"label": "Phosphorylation Score"},
                 square=False)

ax.set_xlabel("Kinase", fontsize=32)
ax.set_ylabel("Residue Position", fontsize=32)
ax.set_title("", fontsize=28, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=20)
plt.yticks(rotation=0, fontsize=20)
plt.tight_layout()

plt.savefig("phosphorylation_heatmap_full.png", dpi=300, bbox_inches='tight')
plt.close()
```
